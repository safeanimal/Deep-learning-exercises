## Utils

```python
# 将一个PIL图像Resize后并转换为Tensor
def resize2tensor(image: Image, size: int) -> Tensor:
    transform = tvt.Compose([
        tvt.Resize(size, interpolation=tvf.InterpolationMode.BICUBIC),
        tvt.ToTensor()
    ])
    return transform(image)


def show_images_tensor(images_per_row: int, images_tensor: Tensor, size=16):
    """
    展示图片
    :param images_per_row: 每行展示的图片数量
    :param images_tensor: B*C*W*H，每个元素取值范围在[-1,1]之间
    :return:
    """

    images_tensor = 0.5 + 0.5 * images_tensor
    # Create a grid of images using make_grid
    grid = tvu.make_grid(images_tensor, nrow=images_per_row, padding=2, normalize=False)

    # plot绘画需要W*H*C的图片，即通道在最后一维
    grid_np = grid.numpy().transpose((1, 2, 0))

    # Display the grid of images using matplotlib
    plt.figure(figsize=(size, size))
    plt.imshow(grid_np)
    plt.axis("off")
    plt.show()
```

## Dataset
要将图像转为值在[-1,1]之间的Tensor

```python
class ImageDataset(Dataset):
    def __init__(self, img_size, img_folder, min_size, img_nums):
        super().__init__()
        self.all_img_names = sorted(os.listdir(img_folder))
        self.img_folder = img_folder
        
        self.img_names = []
        # 将低像素的图片过滤掉
        for img_name in self.all_img_names:
            if len(self.img_names) >= img_nums:
                break
            width, height = Image.open(os.path.join(self.img_folder, img_name)).size
            if width >= min_size and height >= min_size:
                self.img_names.append(img_name)

        self.input_transform = transforms.Compose([
            transforms.Resize((img_size, img_size),
                              interpolation=tvf.InterpolationMode.BICUBIC),
            transforms.ToTensor(),
            transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)) # 使用imagenet的mean和std将值转化到[-1, 1]之间
        ])

    def __getitem__(self, item):
        img_name = self.all_img_names[item]
        img_path = os.path.join(self.img_folder, img_name)
        img = Image.open(img_path).convert('RGB')

        input_img = self.input_transform(img)

        return input_img

    def __len__(self):
        return len(self.img_names)
```

总步数n_steps=1000，$\beta\in[10^{-4},0.02]$且为线性间隔取值

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

n_steps = 500
# 注意下标从0开始
beta = torch.linspace(0.0001, 0.02, n_steps).to(device)
alpha = 1 - beta
alpha_bar = torch.cumprod(alpha, dim=0)
sigma = torch.sqrt(beta)
```

```python
print('beta[0:10]:', beta[0:10])
print('alpha[0:10]:', alpha[0:10])
print('alpha_bar[0:10]', alpha_bar[0:10])
print('sigma[0:10]', sigma[0:10])
```

### Get $x_t$
$$\boldsymbol{x_t}=\sqrt{\bar{\alpha}_t}\boldsymbol{x_0}+\sqrt{1-\bar{\alpha}_t}\boldsymbol{\epsilon}$$
$$\boldsymbol{\epsilon}\sim N(0,I)$$
对于`get_xt_given_x0_and_t`，每次使用的$\boldsymbol{\epsilon}$是一样的，应该从外部传入

```python
def get_xt_given_x0_and_t(x0, t, eps):
    """
        x0: (batch_size, C, W, H)
        t: (batch_size,)
    """
    sqrt_alpha_bar_t = torch.sqrt(alpha_bar[t-1]).view(x0.shape[0], 1, 1, 1)
    sqrt_one_minus_sqrt_alpha_bar_t = torch.sqrt(1-alpha_bar[t-1]).view(x0.shape[0], 1, 1, 1)

    mean = sqrt_alpha_bar_t * x0
    std = sqrt_one_minus_sqrt_alpha_bar_t
    xt = mean + std * eps
    return xt
```

### Sampling with trained model
$$\boldsymbol{x}_{t-1}=\frac{1}{\sqrt{\alpha}}(\boldsymbol{x}_t-\frac{1-\alpha_t}{\sqrt{1-\bar{\alpha}_t}}*\boldsymbol{\epsilon}_\theta(\boldsymbol{x}_t, t))+\sigma_t*\boldsymbol{z}$$
$$\boldsymbol{z}\sim N(0,I)$$

```python
def sampling(xt, T, model, start, end, step):
    """
        xt: (C, W, H)
    """
    xt = xt.view(1, *xt.shape) # (1, C, W, H)
    sampling_results = []
    # 计算x_{T-1},...,x0
    for i in range(T, 0, -1):
        z = torch.randn(*xt.shape).to(device) # (1, C, W, H)

        t_for_model = (torch.zeros((xt.shape[0],)).to(device) + i) # (1, )

        eps_theta = model(xt, t_for_model)

        assert not torch.any(torch.isnan(eps_theta)), 'in function `sampling`, eps_theta is nan!'

        coeff = 1. / torch.sqrt(alpha[i-1])
        coeff_eps_theta = (1-alpha[i-1]) / torch.sqrt(1 - alpha_bar[i-1])
        if i == 1:
            xt = coeff * (xt - coeff_eps_theta * eps_theta)
        else:
            xt = coeff * (xt - coeff_eps_theta * eps_theta) + sigma[i-1] * z
        
        # 存放结果，在[start, end]中每step取一个样本，最后一个样本也加进去
        if i >= start and i <= end and (i-1) % step == 0 or i == end:
            sampling_results.append(xt.view(3, img_size, img_size))
    sampling_results = torch.stack(sampling_results)

    return sampling_results
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self):
        super(PositionalEncoding, self).__init__()
        
    def forward(self, X, t):
        """
            X: (batch size, 3, img_size, img_size)
            t: (batch size,)
        """
        t = t.view(X.shape[0], 1, 1, 1) # t: (batch size, 1, 1, 1)
        P = (torch.sin(t)/t).to(X.device)
        X = X + P
        return X
```

```python
class EpsilonPredictor(nn.Module):
    def __init__(self):
        super(EpsilonPredictor, self).__init__()
        # 卷积后的大小：output_size = (n + 2p - k) / s + 1
        # 转置卷积后的大小：output_size = (n - 1) * s - 2 * p + k + out_padding
        
        # 输入 n=32
        # 输出 16 = (32 + 2p - k)/s + 1
        # 取p=1, k=4, s=2满足条件
        self.conv1 = nn.Conv2d(3, 16, 4, 2, 1)
        # 输入 n=16
        # 输出 8 = (16 + 2p - k)/s + 1
        # 取p=1 k=4, s=2满足条件
        self.conv2 = nn.Conv2d(16, 32, 4, 2, 1)
        
        self.trans_conv1 = nn.ConvTranspose2d(32)
        
        
        self.positional_encoding = PositionalEncoding()
        
    def forward(self, x, t):
        x = self.positional_encoding(x, t)
        for layer in self.layers:
            x = layer(x)
            assert not torch.isnan(x).any(), "NaN values found in layer output"
        return x
```

```python
class Unet(nn.Module):
    def __init__(self):
        super(Unet, self).__init__()
        self.left_layers1 = nn.Sequential(
            nn.Conv2d(1, 64, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 0),
            nn.ReLU()
        )
        
        self.left_layers2 = nn.Sequential(
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, 1, 0),
            nn.ReLU()
        )
        
        self.left_layers3 = nn.Sequential(
            nn.MaxPool2d(2),
            nn.Conv2d(128, 256, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(256, 256, 3, 1, 0),
            nn.ReLU()
        )
        
        self.left_layers4 = nn.Sequential(
            nn.MaxPool2d(2),
            nn.Conv2d(256, 512, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(512, 512, 3, 1, 0),
            nn.ReLU()
        )
        
        self.bottom_layers = nn.Sequential(
            nn.MaxPool2d(2),
            nn.Conv2d(512, 1024, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(1024, 1024, 3, 1, 0),
            nn.ReLU()
        )
        # As for ConvTransposed2d, output_size = (input_size - 1) * stride - 2 * padding + kernel_size + out_padding
        # 
        self.up_conv4 = nn.ConvTranspose2d(1024, 512, 2)
        
        self.right_layers4 = nn.Sequential(
            nn.Conv2d(1024, 512, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(512, 512, 3, 1, 0),
            nn.ReLU()
        )
        
        self.up_conv3 = nn.ConvTranspose2d(512, 256, 2)
        
        self.right_layers3 = nn.Sequential(
            nn.Conv2d(512, 256, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(256, 256, 3, 1, 0),
            nn.ReLU()
        )
        
        self.up_conv2 = nn.ConvTranspose2d(256, 128, 2)
        
        self.right_layers2 = nn.Sequential(
            nn.Conv2d(256, 128, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, 1, 0),
            nn.ReLU()
        )
        
        self.up_conv1 = nn.ConvTranspose2d(128, 64, 2)
        
        self.right_layers1 = nn.Sequential(
            nn.Conv2d(128, 64, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 0),
            nn.ReLU(),
            nn.Conv2d(64, 2, 1, 1, 0)
        )
        
    def forward(self, x):
        x = nn.Conv2d(3, 1, 3, 1, 1)(x)
        left_output1 = self.left_layers1(x)
        print(left_output1.shape)
        left_output2 = self.left_layers2(left_output1)
        print(left_output2.shape)
        left_output3 = self.left_layers3(left_output2)
        print(left_output3.shape)
        left_output4 = self.left_layers4(left_output3)
        print(left_output4.shape)
        
        bottom_output = self.bottom_layers(left_output4)
        right_output4 = self.up_conv4(bottom_output)
        skip4 = tvf.center_crop(left_output4, output_size=[right_output4.shape[-2], right_output4.shape[-1]])  
        right_input4 = torch.cat((skip4, right_output4), dim=1)
        right_output3 = self.up_conv3(self.right_layers4(right_input4))
        
        skip3 = tvf.center_crop(left_output3, output_size=[right_output3.shape[-2], right_output3.shape[-1]])
        right_input3 = torch.cat((skip3, right_output3), dim=1)
        right_output2 = self.up_conv2(self.right_layers3(right_input3))
        
        skip2 = tvf.center_crop(left_output2, output_size=[right_output2.shape[-2], right_output2.shape[-1]])
        right_input2 = torch.cat((skip2, right_output2), dim=1)
        right_output1 = self.up_conv1(self.right_layers2(right_input2))
        
        skip1 = tvf.center_crop(left_output1, output_size=[right_output1.shape[-2], right_output1.shape[-1]])
        right_input1 = torch.cat((skip1, right_output1), dim=1)
        output = self.right_layers1(right_input1)
        
        return output
```

### Train the model
$$\nabla_{\theta}\Vert \boldsymbol{\epsilon}-\boldsymbol{\epsilon}_{\theta}(x_t, t)\Vert^2$$

```python
model = EpsilonPredictor().to(device)

criterion = nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

epoch_num = 50

model.train() # 开启训练模式！
for i in tqdm(range(epoch_num), desc='Training'):
    for input_images in training_dataloader:
        input_images = input_images.to(device)

        eps = torch.randn(input_images.shape).to(device)

        t = torch.randint(1, n_steps, (input_images.shape[0],)).long().to(device)

        xt = get_xt_given_x0_and_t(input_images, t, eps)

        eps_theta = model(xt, t)

        assert not torch.isnan(eps_theta).any(), "NaN values found in eps_theta"

        assert not torch.isnan(eps).any(), "NaN values found in eps"

        loss = criterion(eps_theta, eps)
        loss.backward()

        assert not torch.isnan(loss), "NaN values found in loss"

        # 打印梯度
        #for name, param in model.named_parameters():
            #if param.grad is not None:
                #print(name, torch.max(param.grad))

        # nn.utils.clip_grad_norm_(model.parameters(), 2, norm_type=2) # 梯度裁剪

        optimizer.step()
        optimizer.zero_grad()
    torch.save(model.state_dict(), "model_temp.pth")
```
