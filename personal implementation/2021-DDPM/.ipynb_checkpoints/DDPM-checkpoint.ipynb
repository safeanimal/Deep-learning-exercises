```python
import torch
from torch import nn
from torch.utils.data import DataLoader
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from tqdm import tqdm
```

## Generate 2d Swiss Roll
`n_samples` is the number of data points.

`noise` is the standard deviation of the Gaussian noise.

`t` is a 2D array with shape (1, n_samples). It is filled with random numbers uniformly distributed in the range (3π, 9π).

`x` and `y` are the coordinates of the points in the Swiss roll.

`X` is the final data matrix. It is a 2D array with shape (n_samples, 2), where each row is a point in the Swiss roll.

```python
# Create a 2D Swiss roll
def create_swiss_roll(batch_size=100):
    n_samples = 1000
    noise = 0.5

    swiss_rolls = []
    
    for i in range(batch_size):
        t = 3.0 * np.pi * (1 + 2 * torch.rand(1, n_samples))
        x = t * torch.cos(t)
        y = t * torch.sin(t)

        X = torch.cat((x, y), dim=0)
        X += noise * torch.randn(2, n_samples)
        X = X.T
        swiss_rolls.append(X)

    swiss_rolls = torch.stack(swiss_rolls)  # 将张量列表堆叠为一个张量
    
    return swiss_rolls
```

```python
swiss_rolls = create_swiss_roll(2)
print(swiss_rolls.shape)
# Plot the Swiss roll
plt.scatter(swiss_rolls[0, :, 0], swiss_rolls[0, :, 1])
plt.show()
```

## Denosing Diffusion Probabilistic Model
### Set hyperparameters
$$\beta_t$$
$$\alpha_t = 1 - \beta_t$$
$$\bar{\alpha}=\prod_{s=1}^T\alpha_s$$
$$\sigma_t = \beta_t$$

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

n_steps = 30
beta = torch.linspace(0.001, 0.01, n_steps).to(device)
alpha = 1- beta
alpha_bar = torch.cumprod(alpha, dim=0)
sigma = beta
```

### Get $x_t$
$$x_t=\sqrt{\bar{\alpha}_t}x_0+\sqrt{1-\bar{\alpha}_t}\epsilon$$
$$\epsilon\sim N(0,I)$$

```python
def get_xt_given_x0_and_t(x0, t):
    eps = torch.rand(x0.shape).to(device)
    xt = torch.sqrt(alpha_bar[t]) * x0 + torch.sqrt(1 - alpha_bar[t]) * eps
    return xt, eps
```

### Sampling with trained model
$$x_{t-1}=\frac{1}{\sqrt{\bar{\alpha}}}(x_t-\frac{\beta_t}{\sqrt{1-\bar{\alpha}_t}}*\epsilon_\theta(x_t, t))+\sigma_t*z$$

```python
def sampling(xt, t, model):
    sampling_results = []
    # 计算x_{t-1},...,x0
    for i in range(t-1, -1, -1):
        z = torch.rand(*xt.shape).to(device)
            
        xt = xt.view(-1, 1, 1000, 2)
            
        eps_theta = model(xt, t)
        xt = 1. / torch.sqrt(alpha_bar[i]) * (xt - beta[i] / torch.sqrt(1 - alpha_bar[i]) * eps_theta) + sigma[i] * z
        # 存放结果
        sampling_results.append(xt)
        
    sampling_results = torch.stack(sampling_results)
        
    return sampling_results
```

### Train the model
$$\nabla_{\theta}\Vert \epsilon-\epsilon_{\theta}(x_t, t)\Vert^2$$

```python
def train():
    swiss_rolls = create_swiss_roll(batch_size=200).to(device)
        
    model = EPS().to(device)
        
    criterion = nn.MSELoss()
        
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
        
    epoch_num = 100
    for i in tqdm(range(epoch_num), desc='Training'):
        for x0 in swiss_rolls:
            eps_seq = []
            eps_theta_seq = []
            for t in range(1, n_steps):
                xt, eps = get_xt_given_x0_and_t(x0, t)
                    
                xt = xt.view(-1, 1, 1000, 2)
                    
                eps_theta = model(xt, t)
                eps_seq.append(eps)
                eps_theta_seq.append(eps_theta)

            eps_seq = torch.stack(eps_seq)
            eps_theta_seq = torch.stack(eps_theta_seq)

            loss = criterion(eps_theta_seq, eps_seq)
            loss.backward()

            optimizer.step()
            optimizer.zero_grad()
        
    print('Over!')
        
    noise = torch.rand(*swiss_rolls[0].shape).to(device)
    sampling_results = sampling(noise, n_steps, model)
    return sampling_results
```

```python
class EPS(nn.Module):
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Conv2d(1, 3, 3, 1, 1),
            nn.ReLU(),
            nn.Conv2d(3, 8, 3, 1, 1),
            nn.ReLU(),
            nn.Conv2d(8, 1, 3, 1, 1)
        )
    
    def forward(self, xt, t):
        xt = xt + t
        return self.layers(xt)
```

## Run

```python
sampling_results = train()
```
