```python
import torch
```

```python
length=5
d = 20
```

```python
nu = torch.arange(0, length).unsqueeze(-1)
print(nu, nu.shape)
```

```python
de = torch.pow(1000, torch.arange(0, d, 2) / d).unsqueeze(0) # 1000^(2j/d)
print(de, de.shape)
```

```python
w = nu / de
print(w, w.shape)
```

```python
sinw = torch.sin(w)
cosw = torch.cos(w)
print(sinw, sinw.shape)
print(cosw, cosw.shape)
```

```python
encoding_matrix = torch.zeros(length, d)
print(encoding_matrix, encoding_matrix.shape)
```

```python
encoding_matrix[:,0::2] = sinw
print(encoding_matrix, encoding_matrix.shape)
```

```python
encoding_matrix[:,1::2] = cosw
print(encoding_matrix, encoding_matrix.shape)
```

```python
t = torch.randint(1, length+1, (length,))
print(t, t.shape)
```

```python
output = encoding_matrix[t, ...]
print(output, output.shape)
```
